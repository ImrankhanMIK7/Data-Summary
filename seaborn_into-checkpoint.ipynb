### Visualizing statistical realtionship

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# check data set
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips.shape
```

```python
tips.head()
```

```python
tips.isnull().sum()
```

```python
tips.info
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='sex',data=tips)
```

```python

sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=tips)
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=tips,style='smoker')
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=tips,size='size')
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=tips,col='time')
```

```python
tips['total_bill'].sum()
tips.columns
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
tip = tips.groupby('day')['total_bill'].sum() 
tip
```

```python
tip.plot(kind='bar')
```

```python
sns.relplot(x='day',y='total_bill',data=tips,hue='sex',col='time')
```
